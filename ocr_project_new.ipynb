```python
import os  # İşletim sistemi ile ilgili işlemler için os modülünü içe aktar
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'  # TensorFlow'un detaylı bilgi (info) loglarını kapat, sadece uyarı ve hata mesajlarını göster
import numpy as np  # Sayısal işlemler ve dizi (array) işlemleri için NumPy modülünü içe aktar
import tensorflow as tf  # Derin öğrenme ve model eğitimi için TensorFlow modülünü içe aktar
import cv2  # Görüntü işleme işlemleri için OpenCV modülünü içe aktar
import matplotlib.pyplot as plt  # Grafik çizimleri için Matplotlib'in pyplot modülünü içe aktar
import seaborn as sns  # Gelişmiş görselleştirmeler (örneğin ısı haritaları) için Seaborn modülünü içe aktar
from tensorflow.keras.models import Sequential  # Keras'tan sıralı (sequential) model yapısını içe aktar
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout  # CNN için gerekli katmanları içe aktar
from sklearn.metrics import classification_report, confusion_matrix  # Sınıflandırma raporu ve karışıklık matrisini hesaplamak için fonksiyonları içe aktar
from sklearn.model_selection import train_test_split  # Veri setini eğitim/test bölmek için fonksiyonu içe aktar
from sklearn.manifold import TSNE  # Yüksek boyutlu verileri 2 boyuta indirgemek için t-SNE algoritmasını içe aktar

tf.get_logger().setLevel('ERROR')  # TensorFlow log seviyesini hata (ERROR) seviyesine çekerek gereksiz bilgi mesajlarını kapat

# Sabitler ve ayarlar:
seed = 42  # Tekrarlanabilirlik için rastgelelik sabiti
np.random.seed(seed)  # NumPy için rastgelelik tohumunu ayarla
tf.random.set_seed(seed)  # TensorFlow için rastgelelik tohumunu ayarla
target_size = (28, 28)  # Görüntülerin hedef boyutu (MNIST: 28x28 piksel)
batch_size = 32  # Eğitim sırasında kullanılacak batch (küme) boyutu
num_classes = 10  # Sınıf sayısı (0-9 rakamları)

##############################################
# 1. Veri Setlerinin Oluşturulması
##############################################
# 1.1 Handwriting (El Yazısı) Verisi: MNIST
(x_train_hw, y_train_hw), (x_test_hw, y_test_hw) = tf.keras.datasets.mnist.load_data()
# MNIST verisetini (el yazısı rakamlar) eğitim ve test olarak yükle

# 1.2 Typewriter (Daktilo) Verisinin Sentetik Olarak Oluşturulması
def generate_typewriter_image(label, target_size=(28,28)):
    """
    Belirtilen etiket için beyaz arka plan üzerine daktilo stilinde rakam oluşturur.
    """
    img = np.ones(target_size, dtype=np.uint8) * 255  # 28x28 boyutunda, tüm pikselleri 255 (beyaz) olan bir görüntü oluştur
    text = str(label)  # Etiketi stringe çevir
    font = cv2.FONT_HERSHEY_SIMPLEX  # Yazı tipi olarak basit bir font kullan
    font_scale = 0.8  # Yazının boyutunu ayarla
    thickness = 2  # Yazı kalınlığını ayarla
    text_size = cv2.getTextSize(text, font, font_scale, thickness)[0]  # Yazının piksel cinsinden boyutunu hesapla
    textX = (target_size[1] - text_size[0]) // 2  # Yazıyı yatayda ortalamak için x koordinatını hesapla
    textY = (target_size[0] + text_size[1]) // 2  # Yazıyı dikeyde ortalamak için y koordinatını hesapla
    cv2.putText(img, text, (textX, textY), font, font_scale, (0,), thickness, cv2.LINE_AA)
    # Belirlenen koordinatlara, siyah renkle (0) metni ekle
    return img  # Oluşturulan görüntüyü döndür

# Eğitim etiketlerine karşılık gelen daktilo stilinde görüntüleri oluştur (sentetik veri)
x_train_tp = np.array([generate_typewriter_image(label, target_size) for label in y_train_hw])
y_train_tp = y_train_hw.copy()  # Etiketler aynı olduğu için kopyala
# Test etiketlerine karşılık gelen daktilo stilinde görüntüleri oluştur
x_test_tp = np.array([generate_typewriter_image(label, target_size) for label in y_test_hw])
y_test_tp = y_test_hw.copy()  # Etiketleri kopyala

##############################################
# 2. Eğitim, Doğrulama ve Test Kümelemelerinin Yapılması
##############################################
# El yazısı verisini eğitim ve doğrulama olarak %70/%30 oranında ayır
x_train_hw_train, x_train_hw_val, y_train_hw_train, y_train_hw_val = train_test_split(
    x_train_hw, y_train_hw, test_size=0.3, random_state=seed)
# Daktilo verisini de aynı şekilde ayır
x_train_tp_train, x_train_tp_val, y_train_tp_train, y_train_tp_val = train_test_split(
    x_train_tp, y_train_tp, test_size=0.3, random_state=seed)

# Test verileri için el yazısı ve daktilo verilerini belirle (ayrım yapılmadan kullanılacak)
x_test_hw_final, y_test_hw_final = x_test_hw, y_test_hw
x_test_tp_final, y_test_tp_final = x_test_tp, y_test_tp

# Görüntüler tek kanal (grayscale) olduğu için kanal boyutunu ekle: (28,28) -> (28,28,1)
x_train_hw_train = x_train_hw_train[..., np.newaxis]
x_train_tp_train = x_train_tp_train[..., np.newaxis]
x_train_hw_val = x_train_hw_val[..., np.newaxis]
x_train_tp_val = x_train_tp_val[..., np.newaxis]
x_test_hw_final = x_test_hw_final[..., np.newaxis]
x_test_tp_final = x_test_tp_final[..., np.newaxis]

# Eğitim verilerini el yazısı ve daktilo verilerini birleştirerek oluştur
x_train_combined = np.concatenate([x_train_hw_train, x_train_tp_train], axis=0)
y_train_combined = np.concatenate([y_train_hw_train, y_train_tp_train], axis=0)
# Doğrulama verilerini de birleştir
x_val_combined = np.concatenate([x_train_hw_val, x_train_tp_val], axis=0)
y_val_combined = np.concatenate([y_train_hw_val, y_train_tp_val], axis=0)
# Test verilerini de birleştir
x_test_combined = np.concatenate([x_test_hw_final, x_test_tp_final], axis=0)
y_test_combined = np.concatenate([y_test_hw_final, y_test_tp_final], axis=0)

##############################################
# 3. Görüntü Ön İşleme: OpenCV İşlemleri
##############################################
def cv2_preprocess(image):
    """
    Gelen görüntüyü OpenCV kullanarak ön işler:
      - Gerekirse TensorFlow tensörünü numpy array'e çevirir.
      - Kanal boyutunu ayarlar.
      - Gaussian blur uygular (gürültüyü azaltır).
      - Otsu eşikleme ile binary görüntü elde eder.
      - Canny kenar tespiti uygular.
      - Elde edilen eşik ve kenar görüntülerini normalize eder.
      - Sonuçta 2 kanallı (eşik ve kenar) bir görüntü döndürür.
    """
    if isinstance(image, tf.Tensor):
        image = image.numpy()  # TensorFlow tensörünü numpy array'e çevir
    if image.ndim == 3 and image.shape[-1] == 1:
        image = np.squeeze(image, axis=-1)  # Tek kanal varsa, boyutu sıkıştır (squeeze)
    image_uint8 = np.uint8(image)  # Görüntüyü uint8 formatına çevir (OpenCV gerektirir)
    blurred = cv2.GaussianBlur(image_uint8, (3, 3), 0)  # 3x3 kernel ile Gaussian blur uygula
    # Otsu eşikleme: binary inverse thresholding
    _, thresh = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    thresh_norm = thresh.astype('float32') / 255.0  # Eşiklenmiş görüntüyü [0,1] aralığına normalize et
    edges = cv2.Canny(blurred, 50, 150)  # Canny kenar tespiti uygula (eşik değerleri: 50 ve 150)
    edges_norm = edges.astype('float32') / 255.0  # Kenar görüntüsünü normalize et
    thresh_norm = np.expand_dims(thresh_norm, axis=-1)  # Normalize edilmiş eşik görüntüsüne kanal boyutu ekle
    edges_norm = np.expand_dims(edges_norm, axis=-1)  # Normalize edilmiş kenar görüntüsüne kanal boyutu ekle
    combined = np.concatenate([thresh_norm, edges_norm], axis=-1)  # İki kanalı (eşik ve kenar) birleştir
    return combined  # İşlenmiş görüntüyü döndür

def preprocess_py(image, label):
    # tf.data API ile kullanılabilmesi için OpenCV ön işleme fonksiyonunu tf.py_function ile sarmala
    processed = tf.py_function(func=lambda x: cv2_preprocess(x), inp=[image], Tout=tf.float32)
    processed.set_shape((target_size[0], target_size[1], 2))  # İşlenmiş görüntünün sabit şeklini belirle (28x28, 2 kanal)
    return processed, label  # İşlenmiş görüntüyü ve etiketini döndür

##############################################
# 4. tf.data.Dataset Nesnelerinin Oluşturulması
##############################################
# Eğitim verisini (birleştirilmiş el yazısı ve daktilo) tf.data.Dataset olarak oluştur
train_ds = tf.data.Dataset.from_tensor_slices((x_train_combined, y_train_combined))
train_ds = train_ds.map(preprocess_py, num_parallel_calls=tf.data.AUTOTUNE)  # Her örneğe ön işleme uygula
train_ds = train_ds.batch(batch_size).prefetch(tf.data.AUTOTUNE)  # Batch'le ve veriyi önceden yükle (prefetch)

# Doğrulama verisini oluştur
val_ds = tf.data.Dataset.from_tensor_slices((x_val_combined, y_val_combined))
val_ds = val_ds.map(preprocess_py, num_parallel_calls=tf.data.AUTOTUNE)  # Ön işleme uygula
val_ds = val_ds.batch(batch_size).prefetch(tf.data.AUTOTUNE)  # Batch'le ve prefetch kullan

# Test verisini oluştur
test_ds = tf.data.Dataset.from_tensor_slices((x_test_combined, y_test_combined))
test_ds = test_ds.map(preprocess_py, num_parallel_calls=tf.data.AUTOTUNE)  # Ön işleme uygula
test_ds = test_ds.batch(batch_size).prefetch(tf.data.AUTOTUNE)  # Batch'le ve prefetch ile hızlandır

##############################################
# 5. Modelin Tanımlanması ve Eğitilmesi
##############################################
def build_model(input_shape, num_classes):
    """
    Bir Convolutional Neural Network (CNN) modeli oluşturur.
    Mimarisi:
      - 32 filtreli Conv2D katmanı (ReLU aktivasyonlu)
      - 2x2 boyutunda MaxPooling2D katmanı
      - 64 filtreli Conv2D katmanı (ReLU aktivasyonlu)
      - Tekrar 2x2 MaxPooling2D katmanı
      - Flatten katmanı: 2 boyutlu veriyi 1 boyuta indirger
      - 128 nöronlu Dense katmanı (ReLU aktivasyonlu)
      - Dropout katmanı: Aşırı uyumu önlemek için %50 dropout
      - Çıkış katmanı: Softmax aktivasyonlu, num_classes nöron
    """
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),  # İlk konvolüsyon katmanı: 32 filtre, 3x3 kernel
        MaxPooling2D((2, 2)),  # Havuzlama katmanı: 2x2 boyutunda
        Conv2D(64, (3, 3), activation='relu'),  # İkinci konvolüsyon katmanı: 64 filtre, 3x3 kernel
        MaxPooling2D((2, 2)),  # İkinci havuzlama katmanı
        Flatten(),  # Çok boyutlu çıktıyı tek boyuta çevirir
        Dense(128, activation='relu'),  # Tam bağlantılı katman: 128 nöron, ReLU aktivasyonlu
        Dropout(0.5),  # %50 dropout, aşırı öğrenmeyi (overfitting) engeller
        Dense(num_classes, activation='softmax')  # Çıkış katmanı: Softmax aktivasyonlu, sınıf sayısı kadar nöron
    ])
    return model  # Oluşturulan modeli döndür

input_shape_model = (target_size[0], target_size[1], 2)  # Modelin giriş şekli: 28x28 piksel ve 2 kanal (eşik ve kenar)
model = build_model(input_shape_model, num_classes)  # Modeli oluştur
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
# Modeli derle: Adam optimizasyon, sparse categorical crossentropy kayıp fonksiyonu, doğruluk metriği
model.summary()  # Modelin özetini (katmanları ve parametre sayıları) ekrana yazdır

# Modeli eğitim verisi üzerinde eğit (validation verisi ile birlikte)
epochs = 5  # Eğitim döngüsü (epoch) sayısı
history = model.fit(train_ds, epochs=epochs, validation_data=val_ds, verbose=2)
# Modeli eğit ve eğitim geçmişini (loss, accuracy vb.) sakla

##############################################
# 6. Eğitim Sonuçlarının Görselleştirilmesi
##############################################
plt.figure(figsize=(12, 4))  # Grafik boyutunu ayarla
plt.subplot(1, 2, 1)  # Soldaki subplot: Kayıp (loss) grafiği
plt.plot(history.history['loss'], marker='o', label='Train Loss')  # Eğitim kaybını çiz
plt.plot(history.history['val_loss'], marker='o', label='Val Loss')  # Doğrulama kaybını çiz
plt.title('Loss per Epoch')  # Grafik başlığı
plt.xlabel('Epoch')  # X ekseni etiketi
plt.ylabel('Loss')  # Y ekseni etiketi
plt.legend()  # Açıklama kutusunu göster
plt.grid(True)  # Izgara çizgilerini ekle

plt.subplot(1, 2, 2)  # Sağdaki subplot: Doğruluk (accuracy) grafiği
plt.plot(history.history['accuracy'], marker='o', label='Train Accuracy')  # Eğitim doğruluğunu çiz
plt.plot(history.history['val_accuracy'], marker='o', label='Val Accuracy')  # Doğrulama doğruluğunu çiz
plt.title('Accuracy per Epoch')  # Grafik başlığı
plt.xlabel('Epoch')  # X ekseni etiketi
plt.ylabel('Accuracy')  # Y ekseni etiketi
plt.legend()  # Açıklama kutusunu göster
plt.grid(True)  # Izgara çizgilerini ekle

plt.tight_layout()  # Alt grafiklerin birbirine girmemesini sağlar
plt.show()  # Grafik penceresini aç

##############################################
# 7. Genel Test Sonuçlarının Değerlendirilmesi
##############################################
test_loss, test_acc = model.evaluate(test_ds, verbose=0)  # Test verisi üzerinde modelin performansını değerlendir
print("Test Loss:", test_loss)  # Test kaybını ekrana yazdır
print("Test Accuracy:", test_acc)  # Test doğruluğunu ekrana yazdır

# Test verisi üzerinden tahmin yapıp, gerçek ve tahmin edilen etiketleri topla
y_true = []  # Gerçek etiketleri saklamak için boş liste oluştur
y_pred = []  # Tahmin edilen etiketleri saklamak için boş liste oluştur
for images, labels in test_ds:  # Test datasetindeki her batch için döngü
    preds = model.predict(images, verbose=0)  # Batch üzerindeki tahminleri al
    y_true.extend(labels.numpy())  # Gerçek etiketleri listeye ekle
    y_pred.extend(np.argmax(preds, axis=1))  # En yüksek olasılığa sahip sınıfı (tahmini) listeye ekle
print("General Test Classification Report:")  # Genel test raporu başlığını yazdır
print(classification_report(y_true, y_pred, digits=4))  # Sınıflandırma raporunu (precision, recall, f1-score) yazdır

# Genel test için karışıklık (confusion) matrisini hesapla ve görselleştir
cm = confusion_matrix(y_true, y_pred)  # Karışıklık matrisini hesapla
plt.figure(figsize=(8, 6))  # Görsel boyutunu ayarla
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=[str(i) for i in range(num_classes)],
            yticklabels=[str(i) for i in range(num_classes)])
# Karışıklık matrisini ısı haritası olarak çiz, sınıf etiketleriyle birlikte
plt.xlabel("Predicted")  # X ekseni etiketini belirle
plt.ylabel("True")  # Y ekseni etiketini belirle
plt.title("Confusion Matrix - General Test")  # Grafik başlığını ayarla
plt.show()  # Grafiği göster

##############################################
# 8. Ayrı Değerlendirme: Handwriting vs. Typewriter
##############################################
# El yazısı test seti için ayrı değerlendirme yap
test_ds_hw = tf.data.Dataset.from_tensor_slices((x_test_hw_final, y_test_hw_final))
# El yazısı test görüntüleri ve etiketlerinden dataset oluştur
test_ds_hw = test_ds_hw.map(preprocess_py, num_parallel_calls=tf.data.AUTOTUNE)  # Ön işleme uygula
test_ds_hw = test_ds_hw.batch(batch_size)  # Batch'le
y_true_hw = []  # El yazısı için gerçek etiketler listesi
y_pred_hw = []  # El yazısı için tahmin edilen etiketler listesi
for images, labels in test_ds_hw:  # Dataset üzerinde döngü
    preds = model.predict(images, verbose=0)  # Tahminleri al
    y_true_hw.extend(labels.numpy())  # Gerçek etiketleri ekle
    y_pred_hw.extend(np.argmax(preds, axis=1))  # Tahmin edilen sınıfları ekle
print("Handwriting Test Classification Report:")  # Başlık yazdır
print(classification_report(y_true_hw, y_pred_hw, digits=4))  # El yazısı test raporunu yazdır
cm_hw = confusion_matrix(y_true_hw, y_pred_hw)  # El yazısı için karışıklık matrisini hesapla
plt.figure(figsize=(8, 6))  # Görsel boyutunu ayarla
sns.heatmap(cm_hw, annot=True, fmt='d', cmap='Greens',
            xticklabels=[str(i) for i in range(num_classes)],
            yticklabels=[str(i) for i in range(num_classes)])
# El yazısı karışıklık matrisini ısı haritası olarak çiz
plt.xlabel("Predicted")  # X ekseni etiketini belirle
plt.ylabel("True")  # Y ekseni etiketini belirle
plt.title("Confusion Matrix - Handwriting Test")  # Grafik başlığını belirle
plt.show()  # Grafiği göster

# Daktilo test seti için ayrı değerlendirme yap
test_ds_tp = tf.data.Dataset.from_tensor_slices((x_test_tp_final, y_test_tp_final))
# Daktilo test görüntüleri ve etiketlerinden dataset oluştur
test_ds_tp = test_ds_tp.map(preprocess_py, num_parallel_calls=tf.data.AUTOTUNE)  # Ön işleme uygula
test_ds_tp = test_ds_tp.batch(batch_size)  # Batch'le
y_true_tp = []  # Daktilo için gerçek etiketler listesi
y_pred_tp = []  # Daktilo için tahmin edilen etiketler listesi
for images, labels in test_ds_tp:  # Dataset üzerinde döngü
    preds = model.predict(images, verbose=0)  # Tahminleri al
    y_true_tp.extend(labels.numpy())  # Gerçek etiketleri ekle
    y_pred_tp.extend(np.argmax(preds, axis=1))  # Tahmin edilen sınıfları ekle
print("Typewriter Test Classification Report:")  # Başlık yazdır
print(classification_report(y_true_tp, y_pred_tp, digits=4))  # Daktilo test raporunu yazdır
cm_tp = confusion_matrix(y_true_tp, y_pred_tp)  # Daktilo için karışıklık matrisini hesapla
plt.figure(figsize=(8, 6))  # Görsel boyutunu ayarla
sns.heatmap(cm_tp, annot=True, fmt='d', cmap='Oranges',
            xticklabels=[str(i) for i in range(num_classes)],
            yticklabels=[str(i) for i in range(num_classes)])
# Daktilo karışıklık matrisini ısı haritası olarak çiz
plt.xlabel("Predicted")  # X ekseni etiketini belirle
plt.ylabel("True")  # Y ekseni etiketini belirle
plt.title("Confusion Matrix - Typewriter Test")  # Grafik başlığını belirle
plt.show()  # Grafiği göster

##############################################
# 9. Ek Profesyonel Görselleştirmeler
##############################################
# 9.1 t-SNE Görselleştirmesi (Özellik Uzayının 2 Boyuta İndirgenmesi)
# Modelin ilk katmanından (giriş) Flatten katmanına kadar olan kısmı alarak özellik çıkarım modeli oluştur
feature_model = tf.keras.Model(inputs=model.layers[0].input, outputs=model.layers[5].output)
features = feature_model.predict(test_ds, verbose=0)  # Genel test seti için özellikleri çıkar
tsne = TSNE(n_components=2, random_state=seed, perplexity=30)  # t-SNE algoritmasını 2 boyut ve belirli perplexity değeri ile başlat
tsne_results = tsne.fit_transform(features)  # Özellikleri 2 boyuta indirgeme işlemini uygula
plt.figure(figsize=(8, 6))  # Görsel boyutunu ayarla
scatter = plt.scatter(tsne_results[:, 0], tsne_results[:, 1], c=y_true, cmap='tab10', s=10)
# t-SNE sonuçlarını scatter plot olarak çiz, renkleri gerçek etiketlere göre belirle
plt.colorbar(scatter, ticks=range(num_classes))  # Renk skalasını ekle (sınıf etiketleri)
plt.title("t-SNE Visualization - General Test Features")  # Grafik başlığını belirle
plt.xlabel("t-SNE Component 1")  # X ekseni etiketini belirle
plt.ylabel("t-SNE Component 2")  # Y ekseni etiketini belirle
plt.grid(True)  # Izgara çizgilerini ekle
plt.show()  # Grafiği göster

# Handwriting (El Yazısı) test seti için t-SNE görselleştirmesi
features_hw = feature_model.predict(test_ds_hw, verbose=0)  # El yazısı test seti özelliklerini çıkar
tsne_hw = TSNE(n_components=2, random_state=seed, perplexity=30)  # t-SNE'yi başlat
tsne_results_hw = tsne_hw.fit_transform(features_hw)  # Özellikleri 2 boyuta indirgeme işlemini uygula
plt.figure(figsize=(8, 6))  # Görsel boyutunu ayarla
scatter = plt.scatter(tsne_results_hw[:, 0], tsne_results_hw[:, 1], c=y_true_hw, cmap='tab10', s=10)
# El yazısı özelliklerinin t-SNE scatter plot'unu çiz
plt.colorbar(scatter, ticks=range(num_classes))  # Renk skalasını ekle
plt.title("t-SNE Visualization - Handwriting Test Features")  # Grafik başlığını belirle
plt.xlabel("t-SNE Component 1")  # X ekseni etiketini belirle
plt.ylabel("t-SNE Component 2")  # Y ekseni etiketini belirle
plt.grid(True)  # Izgara çizgilerini ekle
plt.show()  # Grafiği göster

# Typewriter (Daktilo) test seti için t-SNE görselleştirmesi
features_tp = feature_model.predict(test_ds_tp, verbose=0)  # Daktilo test seti özelliklerini çıkar
tsne_tp = TSNE(n_components=2, random_state=seed, perplexity=30)  # t-SNE algoritmasını başlat
tsne_results_tp = tsne_tp.fit_transform(features_tp)  # Özellikleri 2 boyuta indirgeme işlemini uygula
plt.figure(figsize=(8, 6))  # Görsel boyutunu ayarla
scatter = plt.scatter(tsne_results_tp[:, 0], tsne_results_tp[:, 1], c=y_true_tp, cmap='tab10', s=10)
# Daktilo özelliklerinin t-SNE scatter plot'unu çiz
plt.colorbar(scatter, ticks=range(num_classes))  # Renk skalasını ekle
plt.title("t-SNE Visualization - Typewriter Test Features")  # Grafik başlığını belirle
plt.xlabel("t-SNE Component 1")  # X ekseni etiketini belirle
plt.ylabel("t-SNE Component 2")  # Y ekseni etiketini belirle
plt.grid(True)  # Izgara çizgilerini ekle
plt.show()  # Grafiği göster

# 9.2 Sınıf Bazlı Metrik Bar Grafiği (Precision, Recall, F1-score)
report_dict = classification_report(y_true, y_pred, output_dict=True)
# Genel test raporunu sözlük formatında al
classes = [str(i) for i in range(num_classes)]  # Sınıf etiketlerini string olarak hazırla
precision_vals = [report_dict[str(i)]['precision'] for i in range(num_classes)]  # Her sınıf için precision değerlerini çıkar
recall_vals = [report_dict[str(i)]['recall'] for i in range(num_classes)]  # Her sınıf için recall değerlerini çıkar
f1_vals = [report_dict[str(i)]['f1-score'] for i in range(num_classes)]  # Her sınıf için f1-score değerlerini çıkar
plt.figure(figsize=(10,6))  # Bar grafiği için görsel boyutunu ayarla
x_axis = np.arange(num_classes)  # X ekseni için sınıf indekslerini oluştur
width = 0.25  # Barların genişliğini ayarla
plt.bar(x_axis - width, precision_vals, width, label='Precision')  # Precision barlarını çiz (sola kaydırarak)
plt.bar(x_axis, recall_vals, width, label='Recall')  # Recall barlarını çiz (ortalanmış)
plt.bar(x_axis + width, f1_vals, width, label='F1-score')  # F1-score barlarını çiz (sağa kaydırarak)
plt.xticks(x_axis, classes)  # X ekseni etiketlerini sınıf isimleri olarak ayarla
plt.xlabel('Class')  # X ekseni etiketini belirle
plt.ylabel('Score')  # Y ekseni etiketini belirle
plt.title('Classification Metrics per Class')  # Grafik başlığını belirle
plt.legend()  # Açıklama kutusunu ekle
plt.grid(True)  # Izgara çizgilerini ekle
plt.show()  # Grafiği göster

# 9.3 Rastgele Test Örneklerinin Görselleştirilmesi
all_images = []  # Tüm test görüntülerini saklamak için boş liste
all_labels = []  # Tüm test etiketlerini saklamak için boş liste
for images, labels in test_ds:  # Test datasetindeki her batch için döngü
    for i in range(images.shape[0]):  # Batch içindeki her görüntü için döngü
        all_images.append(images[i])  # Görüntüyü listeye ekle
        all_labels.append(labels.numpy()[i])  # Etiketi (numpy formatında) listeye ekle
all_images = np.array(all_images)  # Listeyi numpy array'e çevir
all_labels = np.array(all_labels)  # Etiket listesini numpy array'e çevir
preds = model.predict(test_ds, verbose=0)  # Tüm test verisi üzerinde tahmin yap
all_preds = np.argmax(preds, axis=1)  # Her örnek için en yüksek olasılığa sahip sınıfı al
n_display = 12  # Görüntülenecek örnek sayısı
indices = np.random.choice(len(all_images), n_display, replace=False)  # Rastgele 12 örneğin indeksini seç
plt.figure(figsize=(12, 8))  # Görsel boyutunu ayarla
for j, idx in enumerate(indices):  # Seçilen örnekler üzerinde döngü
    img = all_images[idx].numpy() if isinstance(all_images[idx], tf.Tensor) else all_images[idx]
    # Eğer tensör ise numpy array'e çevir, değilse direkt kullan
    img_disp = img[:, :, 0]  # Görüntünün ilk kanalını (orijinal grayscale görüntü) seç
    true_label = all_labels[idx]  # Gerçek etiketi al
    pred_label = all_preds[idx]  # Tahmin edilen etiketi al
    plt.subplot(3, 4, j+1)  # 3x4 ızgara içinde alt grafik oluştur
    plt.imshow(img_disp, cmap='gray')  # Görüntüyü grayscale olarak göster
    plt.title(f"True: {true_label}, Pred: {pred_label}")  # Alt grafik başlığı: Gerçek ve tahmin edilen etiketler
    plt.axis('off')  # Eksenleri kapat
plt.suptitle("Random Test Examples")  # Genel başlık ekle
plt.tight_layout()  # Grafiklerin düzenli yerleşmesini sağla
plt.show()  # Grafiği göster

```
